## Edge-Level-Feature Engineering

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the Karate Club graph
G = nx.karate_club_graph()
G
```

```python
# Print the number of nodes and edges
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Print the edges
print("Edges:")
for edge in G.edges():
    print(edge)
```

```python
# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500)
plt.title("Karate Club Graph")
plt.show()
```

## Shortest Path Length

```python
# Calculate shortest path lengths for all pairs of nodes
shortest_path_lengths = dict(nx.all_pairs_shortest_path_length(G))

# Display the shortest path lengths
for source_node in G.nodes():
    for target_node in G.nodes():
        if source_node != target_node:
            length = shortest_path_lengths[source_node][target_node]
            if length==2:
               print(f"Shortest path length between Node {source_node} and Node {target_node}: {length}")
```

## Common Neighbors

```python
# Calculate common neighbors for all pairs of nodes
common_neighbors = {}
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 != node2:
            common_neighbors[(node1, node2)] = len(list(nx.common_neighbors(G, node1, node2)))

# Display the number of common neighbors
for (source_node, target_node), count in common_neighbors.items():
    print(f"Common neighbors between Node {source_node} and Node {target_node}: {count}")
```

## Jaccard's Coefficient

```python
# Calculate Jaccard coefficients for all pairs of nodes
jaccard_coefficients = {}
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 != node2:
            common_neighbors = list(nx.common_neighbors(G, node1, node2))
            neighbors_node1 = list(nx.neighbors(G, node1))
            neighbors_node2 = list(nx.neighbors(G, node2))
            union_neighbors = len(neighbors_node1) + len(neighbors_node2) - len(common_neighbors)

            if union_neighbors > 0:
                jaccard_coefficient = len(common_neighbors) / union_neighbors
            else:
                jaccard_coefficient = 0.0

            jaccard_coefficients[(node1, node2)] = jaccard_coefficient

# Display the Jaccard coefficients
for (source_node, target_node), coefficient in jaccard_coefficients.items():
    print(f"Jaccard coefficient between Node {source_node} and Node {target_node}: {coefficient}")
```

## Adamic-Adar Index

```python
# Calculate Adamic-Adar Index for all pairs of nodes
adamic_adar_indices = {}
for node1 in G.nodes():
    for node2 in G.nodes():
        if node1 != node2:
            adamic_adar_index = 0.0
            common_neighbors = list(nx.common_neighbors(G, node1, node2))
            for common_neighbor in common_neighbors:
                degree_common_neighbor = G.degree(common_neighbor)
                adamic_adar_index += 1 / (degree_common_neighbor * 1.0)

            adamic_adar_indices[(node1, node2)] = adamic_adar_index

# Display the Adamic-Adar Indices
for (source_node, target_node), index in adamic_adar_indices.items():
    print(f"Adamic-Adar Index between Node {source_node} and Node {target_node}: {index}")
```
