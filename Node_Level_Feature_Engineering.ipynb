## Node-Level-Feature Engineering

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the Karate Club graph
G = nx.karate_club_graph()
G
```

```python
# Print the number of nodes and edges
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Print the edges
print("Edges:")
for edge in G.edges():
    print(edge)
```

```python
# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=500)
plt.title("Karate Club Graph")
plt.show()
```

## Node Degree

```python
# Compute and print the degree os each node
for node in G.nodes():
  print(f"Degree of {node} is: {G.degree(node)}")
```

```python
from matplotlib.cm import ScalarMappable

degree = nx.degree(G)
# Create a layout for the graph
pos = nx.spring_layout(G)

# Create a figure and axis for the plot
plt.figure(figsize=(15, 10))
plt.axis('off')

# Draw the graph with nodes colored based on degree centrality
node_colors = [degree[node] for node in G.nodes()]
nodes = nx.draw_networkx_nodes(G, pos, node_color=node_colors, cmap=plt.cm.viridis, node_size=200)

# Draw the edges
nx.draw_networkx_edges(G, pos, alpha=0.5)

# Draw labels for the nodes
nx.draw_networkx_labels(G, pos)

# Create a ScalarMappable object to represent the color bar
sm = ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(min(node_colors), max(node_colors)))
sm.set_array(node_colors)

# Add a color bar for the centrality measures
cbar = plt.colorbar(sm, ax=plt.gca())
cbar.set_label("Node Degree")

# Title
plt.title("Karate Club Network with Node Degree")

# Show the plot
plt.show()
```

## VIsualizing all the Centrality Measures

```python
# Create a figure and axis for the plot
fig, axes = plt.subplots(2, 3, figsize=(20, 15))

# Degree Centrality
degree_centrality = [nx.degree_centrality(G)[node] for node in G.nodes()]
nx.draw(G, pos, node_color=degree_centrality, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[0, 0])
axes[0, 0].set_title("Degree Centrality")

# Eigen Vector Centrality
eigenvector_centrality = [nx.eigenvector_centrality(G)[node] for node in G.nodes()]
nx.draw(G, pos, node_color=eigenvector_centrality, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[0, 1])
axes[0, 1].set_title("EigenVector Centrality")

# Betweenness Centrality
betweenness_centrality = [nx.betweenness_centrality(G)[node] for node in G.nodes()]
nx.draw(G, pos, node_color=betweenness_centrality, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[0, 2])
axes[0, 2].set_title("Betweenness Centrality")

# PageRank Centrality
pagerank_centrality = [nx.pagerank(G)[node] for node in G.nodes()]
nx.draw(G, pos, node_color=pagerank_centrality, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[1, 0])
axes[1, 0].set_title("PageRank Centrality")

# Number of Triangles
triangles = [nx.triangles(G)[node] for node in G.nodes()]
nx.draw(G, pos, node_color=triangles, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[1, 1])
axes[1, 1].set_title("Number of Triangles")

# Local Coloring Number (Chromatic Number)
chromatic_number = [nx.coloring.greedy_color(G, strategy="largest_first")[node] for node in G.nodes()]
nx.draw(G, pos, node_color=chromatic_number, cmap=plt.cm.viridis, with_labels=True, node_size=200, ax=axes[1, 2])
axes[1, 2].set_title("Local Coloring Number")

# Adjust subplot layout
plt.tight_layout()

# Show the plots
plt.show()
```

## Clustering Coefficient

```python
# Calculate the clustering coefficient for each node
clustering_coefficients = nx.clustering(G)

# Print the clustering coefficient for each node
for node, coefficient in clustering_coefficients.items():
    print(f"Node {node}: Clustering Coefficient = {coefficient:.4f}")
```

# From Scratch

```python
def local_clustering_coefficient(graph, node):
    neighbors = list(graph.neighbors(node))
    k = len(neighbors)
    if k < 2:
        return 0.0

    total_possible_edges = k * (k - 1) / 2
    actual_edges = 0

    for i in range(k):
        for j in range(i + 1, k):
            if graph.has_edge(neighbors[i], neighbors[j]):
                actual_edges += 1

    clustering_coefficient = (actual_edges) / total_possible_edges
    return clustering_coefficient

for node in G.nodes():
    coefficient = local_clustering_coefficient(G, node)
    print(f"Node {node}: Clustering Coefficient = {coefficient:.4f}")
```
